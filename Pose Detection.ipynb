```python
#Step 1: Import all necessary libraries

import cv2
import mediapipe as mp
```

```python
#Step 2: Identify webcam
#Note that 0 = local webcam; 1 = external webcam

cap = cv2.VideoCapture(0)
```

```python
#Step 6: Leveraging the MediaPipe library used for Pose Detection

mpPose = mp.solutions.pose
pose = mpPose.Pose()
# pose = mpPose.Pose(static_image_mode = False, upper_body_only = False, smooth_landmarks=True, min_detection_confidence=0.5, min_tracking_confidence=0.5)

#Step 7: Draw and connect the landmarks
mpDraw = mp.solutions.drawing_utils
```

```python
#Step 3: Switch on webcam

while True:
    _, img = cap.read()
    
    #Step 8: Convert video/image from BGR to RGB
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    #Step 9: Apply the mediapipe pose detection module for detection
    results = pose.process(imgRGB)
    print(results.pose_landmarks)
    
    #Step 10: Draw landmarks
    if results.pose_landmarks:
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
    
    
    #Step 11: Add a label (optional)
    cv2.putText(img, '10Alytics Pose Detection', (10, 50), cv2.FONT_HERSHEY_PLAIN, 2, (255,255,255), 2)
    
    
    #step 4: show webcam
    cv2.imshow('10Alytics Pose Detection', img)
    if cv2.waitKey(1) & 0xff == ord('q'):
        break
        
#Step 5: Release the capture once all the processing is done.
cap.release()
cv2.destroyAllWindows()
```
